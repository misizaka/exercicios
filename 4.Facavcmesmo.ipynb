```python
#calular a mediana
mediana = sts.median(dataset['Temperatura'])
mediana
```

```python
#substituir
#dataset.loc[(dataset['Temperatura'] <  -130 )  | ( dataset['Temperatura'] >  130), 'Temperatura'] = mediana

# Converter explicitamente a coluna 'Temperatura' para float antes da atribui√ß√£o
dataset['Temperatura'] = dataset['Temperatura'].astype(float)

# Agora voc√™ pode fazer a atribui√ß√£o sem problemas
dataset.loc[(dataset['Temperatura'] < -130) | (dataset['Temperatura'] > 130), 'Temperatura'] = mediana
```
